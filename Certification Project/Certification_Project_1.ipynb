```python
regression_models = [
    LinearRegression(),
    DecisionTreeRegressor(random_state=5),
    RandomForestRegressor(random_state=5)
]
```

```python
def calculate_accuracy_of_model(regmodels, features, label):
    X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.30)
    for model in regmodels:
        model.fit(X_train,y_train)
        prediction = model.predict(X_test)
        name = model.__class__.__name__
        print(f"Root Mean Squared Error of model {0} is {1}",name, np.sqrt(mean_squared_error(prediction,y_test)))
        print('\n')
        print(f"Median absolute Error Error of model {0} is {1}",name, median_absolute_error(prediction,y_test))
        print("*********************************************************************")
```

```python
calculate_accuracy_of_model(regression_models, dftransformedX, label)
```
