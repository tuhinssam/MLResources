## Association Rule Mining

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dfretail = pd.read_excel('Online_Retail.xlsx')
```

```python
dfretail.head()
```

```python
dfretail.sample(10)
```

```python
dfretail.info()
```

```python
dfretail.Country.unique()
```

```python
dfretail['Country'].value_counts()
```

```python
dfretail.describe()
```

```python
dfretail['InvoiceNo']= dfretail['InvoiceNo'].astype('str')
```

```python
 #Get the invoice numbers which doesn't contain 'C'
dfretail['InvoiceNo'] = dfretail[~dfretail['InvoiceNo'].str.contains('C')]
```

```python
dfretail.head()
```

```python
#deleate leading and trailing spaces
dfretail['Description'] = dfretail['Description'].str.strip() 
```

```python
dfretail.head()
```

```python
dfretail.dropna(inplace = True)
```

```python
dfretail['Country'].value_counts()
```

```python
dfretail.info()
```

```python
#create a basket for the data belonginmg to country 'France'
basket = dfretail[dfretail['Country']=='France']
```

```python
basket.sample(10)
```

```python
#sort by InvoiceNo and Description and Get the sum of Quantity
basket = basket.groupby(['InvoiceNo','Description'])['Quantity'].sum()
```

```python
basket
```

```python
basket = basket.unstack()
```

```python
basket.shape
```

```python
basket.head(10)
```

```python
basket = basket.fillna(0)
```

```python
basket.head()
```

```python
def encode(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
basket_sets = basket.applymap(encode)
```

```python
basket_sets.head(10)
```

```python
#we can drop some items which gives very minimum profit
```

```python
frequent_itemset = apriori(basket_sets,min_support=0.07, use_colnames=True) # 7% support
```

```python
frequent_itemset
```

```python
rules = association_rules(frequent_itemset, metric='lift', min_threshold=1)
```

```python
rules.head(10)
```

```python
#antecedents : LHS of Rule
#consequents: RHS of Rule
```

```python
rules[rules['confidence']>=0.8].count()
```
