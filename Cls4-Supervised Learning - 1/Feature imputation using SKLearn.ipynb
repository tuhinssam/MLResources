```python
dfnew['surgery'] = labelEncoder.fit_transform(dfnew['surgery'].astype('str'))
dfnew['age'] = labelEncoder.fit_transform(dfnew['age'].astype('str'))
dfnew['temp_of_extremities'] = labelEncoder.fit_transform(dfnew['temp_of_extremities'].astype('str'))
dfnew['peripheral_pulse'] = labelEncoder.fit_transform(dfnew['peripheral_pulse'].astype('str'))
dfnew['mucous_membrane'] = labelEncoder.fit_transform(dfnew['mucous_membrane'].astype('str'))
dfnew['capillary_refill_time'] = labelEncoder.fit_transform(dfnew['capillary_refill_time'].astype('str'))
dfnew['abdomo_appearance'] = labelEncoder.fit_transform(dfnew['abdomo_appearance'].astype('str'))
dfnew['outcome'] = labelEncoder.fit_transform(dfnew['outcome'].astype('str'))
dfnew['cp_data'] = labelEncoder.fit_transform(dfnew['cp_data'].astype('str'))
```

```python
dfnew['pain'] = labelEncoder.fit_transform(dfnew['pain'].astype('str'))
dfnew['rectal_exam_feces'] = labelEncoder.fit_transform(dfnew['rectal_exam_feces'].astype('str'))
dfnew['abdomen'] = labelEncoder.fit_transform(dfnew['abdomen'].astype('str'))
dfnew['surgical_lesion'] = labelEncoder.fit_transform(dfnew['surgical_lesion'].astype('str'))
```

```python
dfnew['pulse'] = dfnew['pulse'].astype(str).astype(float)
dfnew['rectal_temp'] = dfnew['rectal_temp'].astype(str).astype(float)
dfnew['respiratory_rate'] = dfnew['respiratory_rate'].astype(str).astype(float)
##dfnew['peristalsis'] = dfnew['peristalsis'].astype(str).astype(float)
#dfnew['abdominal_distention'] = dfnew['abdominal_distention'].astype(str).astype(float)
#dfnew['nasogastric_tube'] = dfnew['nasogastric_tube'].astype(str).astype(float)
#dfnew['nasogastric_reflux'] = dfnew['nasogastric_reflux'].astype(str).astype(float)
dfnew['nasogastric_reflux_ph'] = dfnew['nasogastric_reflux_ph'].astype(str).astype(float)
dfnew['packed_cell_volume'] = dfnew['packed_cell_volume'].astype(str).astype(float)
dfnew['total_protein'] = dfnew['total_protein'].astype(str).astype(float)
```

```python
dfnew['peristalsis'] = labelEncoder.fit_transform(dfnew['peristalsis'].astype(str))
dfnew['abdominal_distention'] = labelEncoder.fit_transform(dfnew['abdominal_distention'].astype(str))
dfnew['nasogastric_tube'] = labelEncoder.fit_transform(dfnew['nasogastric_tube'].astype(str))
dfnew['nasogastric_reflux'] = labelEncoder.fit_transform(dfnew['nasogastric_reflux'].astype(str))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
#X = dfnew.drop('outcome',axis = 1)
X = dfnew.drop(['outcome','rectal_temp','respiratory_rate','mucous_membrane','nasogastric_tube','cp_data'],axis = 1)
```

```python
y = dfnew.outcome
```

```python
train_X,test_X,train_y,test_y = train_test_split(X,y, test_size=0.25)
```

```python
dtc = DecisionTreeClassifier(random_state=89)
```

```python
dtc.fit(train_X,train_y)
```

```python
dtcprediced = dtc.predict(test_X)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test_y,dtcprediced)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc =RandomForestClassifier(n_estimators=120)
```

```python
rfc.fit(train_X,train_y)
```

```python
rfcPredict = rfc.predict(test_X)
```

```python
accuracy_score(test_y,rfcPredict)
```
