```python
from PIL import Image
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dogsnparray = np.asarray(Image.open('dogs.jpeg'))
```

```python
dogsnparray.shape
```

```python
dogs2d = dogsnparray.reshape(75480,2)
```

```python
dogs2d.shape
```

```python
dfdogs = pd.DataFrame(dogs2d, columns=['Col1','Col2'])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(dfdogs)
```

```python
kmeans.cluster_centers_
```

```python
dfdogs['cluster'] = kmeans.labels_
```

```python
sns.scatterplot(dfdogs['Col1'], dfdogs['Col2'])
```

```python
sns.lmplot('Col1', 'Col2',data = dfdogs, hue='cluster')
```
